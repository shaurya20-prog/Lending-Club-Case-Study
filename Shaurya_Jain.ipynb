```python
# Dropping columns which has more than 60% of total values are null
lds_2 = lds.dropna(axis='columns',thresh=(0.6* len(lds))) 
```

```python
# drop off columns having 0 values more than 90% in that.
lds_3 = lds_2.loc[:,(lds_2==0).sum() < (0.1*len(lds_2))]
```

```python
# 'loan_amnt', 'funded_amnt', 'funded_amnt_inv' - only loan_amnt can be considered
# 'grade', 'sub_grade' - only grade can be considered
# 'id', 'member_id' - can be dropped since we are not doing analysis for an individual

lds_4 = lds_3.drop(columns = ['funded_amnt', 'funded_amnt_inv','id', 'member_id','application_type',
                              'policy_code','pymnt_plan','initial_list_status'])
```

```python
lds_4.head()
```

```python
lds_4.columns
```

```python
# We can ignore the customer behaviour variables which are basically generated after the loan is approved
# such as delinq_2_years, revol_util, last_credit_pull_d etc.

# Hence for this analysis following are the relevant columns to consider
    # 'loan_amnt', 'term', 'int_rate', 'installment', 'grade', 'emp_title',
    # 'emp_length', 'home_ownership', 'annual_inc', 'verification_status',
    # 'issue_d', 'loan_status', 'purpose', 'addr_state', 'dti'

lds_5 = lds_4.loc[:,['loan_amnt', 'term', 'int_rate', 'installment', 'grade', 'emp_title',
                     'emp_length', 'home_ownership', 'annual_inc', 'verification_status',
                     'issue_d', 'loan_status', 'purpose', 'addr_state', 'dti']]
```

```python
lds_5.head()
```

```python
lds_5.loan_status.value_counts()
```

```python
# We can drop off rows with loan_status as 'Current' since that's not useful for this analysis
lds_5 = lds_5[lds_5.loan_status != 'Current']
lds_5.reset_index()
```

```python
# remove % from interest rate and covert its data type as float
lds_5.int_rate = lds_5.int_rate.apply(lambda x : float(x.replace('%','')))
lds_5.head()
```

```python
lds_5.columns
```

```python
# lds_5['issue_month'] = lds_5.
lds_5[['issue_month','issue_year']] = lds_5.issue_d.str.split('-',expand=True) 
```

```python
lds_5.issue_year = lds_5.issue_year.apply(lambda x : '20'+x)
lds_5.head()
```

```python
#now we can drop issue_d as we have broken it into month and year
lds_5 = lds_5.drop(columns='issue_d')
```

```python
lds_5.head()
```

```python
# method to get numeric value out from a alpha numeric string
def getDigit(x):
    d = '0';
    for i in x:
        if i.isdigit() == True:
            d += i;
    return int(d);
```

```python
lds_5.emp_length = lds_5.emp_length.apply(lambda x : 0 if pd.isnull(x) else getDigit(x))
```

```python
lds_5.emp_length
```

```python
lds_5['num_loan_status'] = lds_5.loan_status.apply(lambda x: 0 if x=='Fully Paid' else 1)
lds_5.head()
```

```python
# lds_5.grade.unique()
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
lds_5['grades_encoded'] = labelencoder.fit_transform(lds_5['grade'])
x=lds_5[['grade','grades_encoded']]
x.groupby(['grade', 'grades_encoded']).size().reset_index(name='Freq')
```

```python
lds_5['grade_mean']=lds_5.groupby(['grades_encoded'])[['num_loan_status']].mean()
lds_5['int_rate_mean']=lds_5.groupby(['int_rate'])[['num_loan_status']].mean()
lds_5['emp_length_mean']=lds_5.groupby(['emp_length'])[['num_loan_status']].mean()
```

```python
# method to get number of missing values in dataframe
def getPercOfMissingValue(df):
    for column in df.columns:
        missValPerc = round(100*df[column].isnull().sum()/len(df), 2);
        if missValPerc > 0:
            print(column, ' -- ', missValPerc,'%');
```

```python
# Let's start with segmented univariate analysis now where target variable is num_loan_status
```

```python
def plotUnivariateAnalysis(column, xLabel, typeVar, order):
    plotType = 'hist'
    if plotType == 'hist':
        plt.figure(figsize=[15,6])
        x = lds_5[lds_5.num_loan_status==0][column];
        y = lds_5[lds_5.num_loan_status==1][column];
        plt.hist([x,y])
        plt.xlabel(xLabel,fontsize=15)
        plt.ylabel('Frequency of Paid/Default Loans',fontsize=15)
        plt.xticks(fontsize=11)
        plt.yticks(fontsize=11)
        plt.show()
        
        if typeVar == 'categorical':
            plt.figure(figsize=[15,6])
            if order == '':
                order = pd.unique(lds_5[column])
            sns.barplot(x=column, y='num_loan_status', data=lds_5, estimator=np.mean, order=order)
            plt.ylabel('norm(no. of Default Loans)',fontsize=15)
            plt.xlabel(xLabel,fontsize=15)
            plt.show()
        
        if typeVar == 'numerical':
            plt.figure(figsize=[15,6])
            sns.boxplot(x='loan_status', y=column, data=lds_5)
            plt.ylabel(xLabel,fontsize=15)
            plt.xlabel('Loan Status',fontsize=15)
            plt.show()

        
```
